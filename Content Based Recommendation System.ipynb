# Movie Recommndation System

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
```

```python
df=pd.read_csv("movies_content.csv")
df.head(6)
```

```python
df=df[['name','description']]
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
df.name=df.name.map(str.lower)
```

```python
tf=TfidfVectorizer()
matrix=tf.fit_transform(df.description).toarray()
```

```python
matrix
```

```python
model=NearestNeighbors(metric='cosine')
model.fit(matrix)
```

```python
try:
    mv=input('Enter Movie Name:').lower()
    mi=df[df.name==mv].index[0]
    dis,idx=model.kneighbors([matrix[mi]],n_neighbors=5)
    print("*********Recommendations for you********")
    for i in idx[0][1:]:
        print(df.loc[i].values[0])
except:
    print("something went wrong,might be due movie not found in database")
```
